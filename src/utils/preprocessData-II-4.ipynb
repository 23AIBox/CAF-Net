```python
# -*- coding: utf-8 -*-
import os
import pickle
from typing import List

import pandas
import numpy as np
from tqdm import tqdm
```

```python
orig_name = 'Dataset II-4.csv'
csv_name = 'II-4'
```

```python
'''
è¯» csv
'''
assert os.path.exists('../../data')
dataset = pandas.read_csv('../../data/Dataset II (mismatch-only)/%s' % orig_name, delimiter=',')
dataset
```

```python
on = dataset['30mer']
off = dataset['30mer_mut']
reads = dataset['GUIDE-SEQ Reads']
reads = np.array(reads)
val = np.zeros(len(reads))
val[reads > 0] = 1
on,off,val
```

```python
tlen =24
ons = []
offs = []
for i in range(len(on)):
    ons.append('-'*(tlen-len(on[i]))+on[i])
for i in range(len(off)):
    offs.append('-'*(tlen-len(off[i]))+off[i])
```

```python
min([len(s) for s in ons])
```

```python
onss = []
for i in tqdm(range(len(ons))):
    seq = []
    for j in range(tlen):
        if ons[i][j]=='N':
            seq.append(offs[i][j])
        else:
            seq.append(ons[i][j])
    onss.append(''.join(seq))
onss[0:10]
```

```python
df = pandas.DataFrame(data={
    'on':onss,
    'off':offs,
    'val':val
})
df.to_csv('../../data/%s.csv' % csv_name, index=False)
```
