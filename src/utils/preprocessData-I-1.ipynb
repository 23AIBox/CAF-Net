```python
# -*- coding: utf-8 -*-
import os
import pickle
from typing import List

import pandas
import numpy as np
from tqdm import tqdm
```

```python
'''
读I-1 csv
'''
assert os.path.exists('../../data')
dataset = pandas.read_csv('../../data/Dataset I (indel&mismatch)/dataset I-1/CIRCLE_seq_10gRNA_wholeDataset.csv', delimiter=',')
dataset
```

```python
on = dataset['sgRNA_seq']
off = dataset['off_seq']
val = dataset['label']
```

```python
tlen =24
for seq in on:
    seq = '-'*(tlen-len(seq))+seq
for seq in off:
    seq = '-'*(tlen-len(seq))+seq
for i in range(len(on)):
    for j in range(tlen):
        if on[i][j]=='N':
            on[i][j]=off[i][j]

df = pandas.DataFrame(data={
    'on':on,
    'off':off,
    'val':val
})
df.to_csv('../../data/I-1.csv',index=False)
```
